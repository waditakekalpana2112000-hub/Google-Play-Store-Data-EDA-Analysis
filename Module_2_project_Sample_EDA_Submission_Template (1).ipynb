## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
#1.creating copy of current database as new_Data_df
New_Data_df=Data_df.copy()
New_Data_df
````

````python
#2.Finding null values
New_Data_df.isnull().sum()
````

````python
#3.converting null values into percentage.
New_Data_df.isnull().sum()/len(New_Data_df)*100
````

````python
#4.Checking the skewness of the rating column.
Skewness = New_Data_df['Rating'].skew()
Skewness #negative skew indicates left-skewed distribution
````

````python
#5.Droping rows with missing values in review or the sentiment review column because these makes EDA process complex.
#for ensuring data quality i would like to remove incomplete records.(by using drop)
New_Data_df.dropna(
    subset=['Translated_Review', 'Sentiment', 'Sentiment_Polarity','Sentiment_Subjectivity'],inplace=True)
````

````python
#6.drop duplicate rows if any
New_Data_df.drop_duplicates(inplace=True)

#7.reset index
New_Data_df.reset_index(drop=True,inplace=True)
````

````python
#8.now we are checking if there is any duplicate row.
New_Data_df.duplicated().sum()
````

````python
#9.The Installs column contains strings with '+' and ',' (for example, "500,000+").
#These non-numeric characters prevent mathematical analysis.
#We need to remove the '+' and ',' to clean the data and then convert the column into integers.
New_Data_df['Installs']=New_Data_df['Installs'].astype(str)
New_Data_df['Installs']=New_Data_df['Installs'].astype(str).str.replace(r'[+,]','',regex=True).astype(int)
````

````python
#10.The reviews column which is stored in object means a string, even they represent the numbers.
#so we have to convert integer type that allows to perform  numerical operation properly.
New_Data_df['Reviews']=New_Data_df['Reviews'].astype(int)
````

````python
#11.convert the price column from string to float.
New_Data_df['Price']=New_Data_df['Price'].astype(str).str.replace('$', '', regex=False).astype(float)
````

````python
#12.clean and convert the Size column into float.
#Step 1: Drop unwanted rows where size is 'Varies with device' or  missing
New_Data_df= New_Data_df[New_Data_df['Size'].notna() & (New_Data_df['Size'] != 'Varies with device')].copy()
#step 2:converting remaining size values to float(MB)
def size_to_mb(size):
  size = str(size).strip()
  if 'M' in size:
    return float(size.replace('M',''))
  elif 'k' in size or 'K' in size:
    return float(size.replace('k','').replace('K','')) / 1024
  else:
    return float(size)

New_Data_df['Size'] = New_Data_df['Size'].apply(size_to_mb)
#step3:convert data type to float
New_Data_df['Size'] =New_Data_df['Size'].astype(float)
````

````python
#13.Convert Last Updated data type from string to Datetime
New_Data_df['Last Updated'] = pd.to_datetime(New_Data_df['Last Updated'])
````

````python
#14.Convert Sentiment columns to float
New_Data_df['Sentiment_Polarity'] = New_Data_df['Sentiment_Polarity'].astype(float)
New_Data_df['Sentiment_Subjectivity'] = New_Data_df['Sentiment_Subjectivity'].astype(float)
````

````python
#15.Extract Year and month from last update
New_Data_df['Last_Updated_Year'] = New_Data_df['Last Updated'].dt.year
````

### What all manipulations have you done and insights you found?

I first created a copy of the original dataset. Then, I handled missing values: for the 'Rating' column, I checked the skewness. then I dropped several columns—including 'Translated_Review', 'Sentiment', 'Sentiment_Polarity', and 'Sentiment_Subjectivity'—because they had too many missing values, which can make EDA process complex. After that, I reset the index so the dataset remained continuous. Next, I performed data type conversions: I converted the 'Installs' column from a string (with non-numeric characters) to integers, the 'Review' column from string to integer, and the 'Price' column from string to float. I also converted the 'Last Updated' column from string to datetime. Finally, I converted the sentiment columns to floats and extracted both the year and month from the 'Last Updated' column.
#insights
1.Missing Data: The rating column was safely filled using the median, and rows with missing sentiment data were dropped to maintain the integrity of the analysis.

2.Data Consistency: All numeric columns—including Installs, Reviews, and Price—were converted to appropriate numerical types, ensuring accurate aggregation and visualization.

3.Time Features: By extracting the year and month from the last update, I enabled trend analysis of app updates over time.

4.Bias Reduction: Duplicate rows were removed, and incomplete sentiment data was dropped, ensuring a more reliable and unbiased exploratory data analysis.

#### Chart - 17 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
# Filter only numeric columns (integers and floats)
numeric_data = New_Data_df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
correlation = numeric_data.corr()

# Set up the figure size
plt.figure(figsize=(10, 8))

# Plot heatmap
sns.heatmap(correlation, annot=True, cmap='rainbow', center=0)
plt.tight_layout()

# Add a title
plt.title('Correlation Heatmap of Numeric Features',fontsize=14,color='green')

# show the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose a heatmap because it clearly shows how the numbers relate to each other. It makes it easy to spot patterns and see which features are connected quickly and clearly.

##### 2. What is/are the insight(s) found from the chart?

1.Growth Factors: Reviews, installs, and size are strongly linked, meaning bigger apps tend to get more installs and reviews.

2.Sentiment Link: Positive and negative sentiment are connected, so more emotional reviews often have stronger opinions.

3.Price Independence: Price doesn’t strongly affect ratings, installs, or reviews, so it’s not a big driver of behavior.

4.Rating Influence: Ratings have only a small connection to installs and reviews, meaning other factors beyond popularity affect ratings.
